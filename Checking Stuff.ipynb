```python
import numpy as np
import os
import time
import PIL.Image as Image


import torch
import torch.nn as nn
import torchvision.transforms as transforms
from torchvision.utils import save_image
import random
```

```python
im = Image.open("d:/Images/dancing.jpg")
arr = np.array(im)
```

```python
XL,XR = 250,310
YU,YD = 200,260
tmp = arr.copy()
tmp[XL:XR,YU:YD,:] = 0
imm = Image.fromarray(tmp)
imm.save("e:/mask.jpg")
```

```python
model = nn.Sequential(nn.Linear(2,8),
                     nn.ReLU(),
                     nn.Linear(8,16),
                     nn.ReLU(),
                     nn.Linear(16,3),
                     nn.Sigmoid())
opt = torch.optim.Adam(model.parameters(),.0002)
criterion = nn.L1Loss()
```

```python
for i in range(5000):
    if i==2000:
        opt = torch.optim.Adam(model.parameters(),.0001)
        
    xs = random.choices(list(range(440)),k=32)
    ys = random.choices(list(range(440)),k=32)
    func = lambda p:p[0]<XL or p[0]>XR or p[1]<YU or p[1]>YD
    ps = list(filter(func,zip(xs,ys)))
    ps = list(map(list,ps))
    
    exout = torch.Tensor([arr[x,y]/255 for x,y in ps])
    
    out = model(torch.Tensor(ps))
    loss = criterion(out,exout)
    
    opt.zero_grad()
    loss.backward()
    opt.step()
    
    print(loss.item())
```

```python
with torch.no_grad():
    ps = [[x,y] for x in range(XL,XR) for y in range(YU,YD)]
    out = model(torch.Tensor(ps))      
```

```python
arr[XL:XR,YU:YD,:] = out.detach().numpy().reshape(XR-XL,-1,3)*255
```
